```python
from numpy import *
import matplotlib.pyplot as plt
```

```python
from scipy.io import loadmat
data = loadmat('../eigenface_demo/cafedata.mat')
```

```python
from sklearn.datasets.base import Bunch
imgs=data['Imgs'].T
imgs-=imgs.min()
imgs/=imgs.max()
rot_imgs=array([img.reshape((60,95)).T for img in imgs])

cafe=Bunch(data=rot_imgs.reshape((80, 60*95)), images=rot_imgs, 
           target=squeeze(data['listIdentities']), gender=squeeze(data['listGenders']),
           DESCR="The CAlifornia Facial Expressions dataset (CAFE)")

import pickle
pickle.dump(cafe, open('cafe.pkl','w'))
```

```python
data = loadmat('../eigenface_demo/pofadata.mat')
from sklearn.datasets.base import Bunch
imgs=data['Imgs'].T
imgs-=imgs.min()
imgs/=imgs.max()
rot_imgs=array([img.reshape((60,73)).T for img in imgs])

pofa=Bunch(data=rot_imgs.reshape((imgs.shape[0], 60*73)), images=rot_imgs, 
           target=squeeze(data['listIdentities']), gender=squeeze(data['listGenders']), expression=squeeze(data['listExpressions']),
           DESCR="Pictures of Facial Affect (POFA)")

import pickle
pickle.dump(pofa, open('pofa.pkl','w'))
```
